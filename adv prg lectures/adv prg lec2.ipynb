```python
#plotting a histogram
import matplotlib.pyplot as plt
df['Age'].plot(kind='hist',bins= 10 , alpha=0.7,) # alpha is for transparency of bars
plt.title
plt.show()
```

```python
#bar plot
import matplotlib.pyplot as plt
#create age bins
bins = [20,30,40,50,60] # define your age range bins
labels = ['20-30', '30-40', '40-50', '50-60']  # defining labels for the bins
df['age_bins'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False) # right = false means the right edge is not included which means 30 is not included in 20-30 bin


# plotting the bar chart for binned age
df['age_bins'].value_counts().plot(kind='bar', alpha=0.7)
plt.title('Age Distribution')
plt.show()
```

```python
#question
# sales data: dana analysis visualization(sales_data/csv)
# load the dataset 
df = pd.read_csv('sales_data.csv')
print(df)
df['Date'] = pd.to_datetime(df['Date']) # converting date column to datetime type
df
```

```python
# calculate the total revenue and units sold for each product in sales_data.csv

total_revenue_units = df.groupby('Product')[['Units Sold', 'Revenue']].sum()
total_revenue_units
```

```python
#create a bar chart showing the units soldfor each product across different dates
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df.pivot(index = 'Date', columns='Product', values='Units Sold').plot(kind='bar',stacked=True) # pivot is used for reshaping the data it is converting unique values from one column into multiple columns

plt.title('Units Sold for each product across diff dates')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()

# if more than one product had sold on same date use stacked = True in plot function it would have been a stacked bar chart
```

```python
# calculate the avg revenue for each product
avg_revenue = df.groupby('Product')['Revenue'].mean()
avg_revenue
```

```python
# students grades : statistical analysis
df = pd.read_csv('Students_Data.csv')
df
```

```python
# calculate the mean grade for each subject
mean_grades = df[['Maths', 'Science', 'English','History']].mean()
mean_grades
columns = ['Maths', 'Science', 'English','History']
```

```python
#identify the student with highest total grade across all subjects
df['total_grade'] = df['Maths'] + df['Science'] + df['English'] + df['History']
max_sorted_list = df.sort_values(by='total_grade', ascending=False)
max_sorted_list = max_sorted_list.iloc[0]
max_sorted_list['Name']
```

```python
df['Total']= df[columns].sum(axis=1)
max_val = df['Total'].idxmax()
top_student = df.loc[max_val]
top_student
```

```python
#create a heatmap showing the correlation between subjects
import seaborn as sns
corr_matrix = df[columns].corr()

# create heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap of Subjects')
plt.show()
```
